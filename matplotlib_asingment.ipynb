Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Scateer plot is a graph that is represented by the scattered discontinious entity. 

```python
import numpy as np 
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("This is the plot")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Ans- When we want to plot multiple plot together in one plot itself, we use the subplot function. It has 3 argument (x,y,z).
X= no of rows in the sub plot
Y= no of column in the subplot and Z= what is the number of the current plot

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.\
Ans- Bar plot is a graph that is used when we have to 1 categorical type of data and 1 numerical type of data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit,color='#19EABA',edgecolor='#327B6A')
plt.xlabel("Profit")
plt.ylabel("Company")
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
Box plot is used to visualise how the data is distributted . It give information about max,min, at data level of different quartile.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(1,2,1)
plt.boxplot(box1)
plt.subplot(1,2,2)
plt.boxplot(box2)
```
